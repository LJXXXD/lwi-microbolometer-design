```python
# PCA and explain variance


# Extracting the substance data (excluding the "Spectra" column) and transpose it
substance_data = data.iloc[:, 1:].T

# Standardizing the data
# scaler = StandardScaler()
# standardized_data = scaler.fit_transform(substance_data)
standardized_data = substance_data - substance_data.mean(axis=0)
# standardized_data = substance_data

# Performing PCA
pca = PCA()
pca_result = pca.fit_transform(standardized_data)


# Explained variance
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()

# Scree plot
plt.figure(figsize=(4, 6))
plt.bar(
    range(1, len(explained_variance) + 1),
    explained_variance,
    alpha=0.5,
    align="center",
    label="Individual explained variance",
)
plt.step(
    range(1, len(cumulative_variance) + 1),
    cumulative_variance,
    where="mid",
    label="Cumulative explained variance",
)

# Set x-axis ticks to be integers
plt.xticks(ticks=range(1, len(explained_variance) + 1))

plt.ylabel("Explained variance ratio")
plt.xlabel("Principal components")
plt.title("Explained Variance by Principal Components")
plt.legend(loc="best")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
# LDA

import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt

# Load the data
file_path = "../data/Test 3 - 4 White Powers/white_powders_modified.xlsx"
data = pd.read_excel(file_path)

# Extract the original spectra and substance data
spectra = data.iloc[:, 0]
substances = data.columns[1:]

# Augment the data with noise
augmented_data = []
augmented_labels = []

for substance in substances:
    original_sample = data[substance].values

    # Add noise to create 10 new samples for each substance
    for i in range(11):  # Including the original sample
        noise = np.random.uniform(-0, 0, size=original_sample.shape) * original_sample
        noisy_sample = original_sample + noise
        augmented_data.append(noisy_sample)
        augmented_labels.append(substance)

# Convert augmented data and labels into DataFrame and Series
augmented_data = np.array(augmented_data)
augmented_labels = np.array(augmented_labels)

# Standardize the data
scaler = StandardScaler()
standardized_data = scaler.fit_transform(augmented_data)

# Perform LDA
lda = LDA(n_components=3)  # We use 2 components for 2D visualization
lda_result = lda.fit_transform(standardized_data, augmented_labels)

# Plotting the LDA results
plt.figure(figsize=(8, 6))
for substance in substances:
    indices = np.where(augmented_labels == substance)
    plt.scatter(lda_result[indices, 0], lda_result[indices, 1], label=substance, alpha=0.7)

plt.xlabel("LD1")
plt.ylabel("LD2")
plt.title("LDA - First Two Linear Discriminants")
plt.legend(loc="best")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
# Plotting the explained variance ratio for each Linear Discriminant
explained_variance_ratio = lda.explained_variance_ratio_

plt.figure(figsize=(8, 6))
plt.bar(
    range(1, len(explained_variance_ratio) + 1),
    explained_variance_ratio,
    alpha=0.5,
    align="center",
    label="Explained Variance Ratio",
)
plt.xlabel("Linear Discriminants")
plt.ylabel("Explained Variance Ratio")
plt.title("Explained Variance by Linear Discriminants")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from itertools import combinations
```

```python
def load_and_clean_data(pathogen_paths, pathogen_names, index_column="Raman Shift"):
    all_pathogen_samples_list = []
    for i, pathogen_path_i in enumerate(pathogen_paths):
        # Load the datasets
        pathogen_i_data = pd.read_excel(pathogen_path_i)

        # Clean column names
        pathogen_i_data.columns = pathogen_i_data.columns.str.strip()

        # Set 'Raman Shift' as the index before transposing to keep it as feature names
        pathogen_i_samples = pathogen_i_data.set_index(index_column).T

        # Assign label of pathogen to the samples
        pathogen_i_samples["Pathogen"] = pathogen_names[i]

        # Assign Concentration
        pathogen_i_samples["Concentration"] = pathogen_i_samples.index

        # Add pathogen_i to list
        all_pathogen_samples_list.append(pathogen_i_samples)

    combined_samples = pd.concat(all_pathogen_samples_list)

    return combined_samples
```

```python
def standardize_and_pca(data):
    # Drop the 'Pathogen' and 'Concentration' columns to isolate features
    features = data.drop(columns=["Pathogen", "Concentration"]).apply(
        pd.to_numeric, errors="coerce"
    )

    # Extract 'Pathogen' and 'Concentration' columns as separate labels
    label_Pathogen = data["Pathogen"]
    label_Concentration = data["Concentration"]

    # Drop any rows with NaN values in the features DataFrame
    features.dropna(inplace=True)

    # Reset indices to ensure clean, sequential indexing
    features.reset_index(drop=True, inplace=True)
    label_Pathogen.reset_index(drop=True, inplace=True)
    label_Concentration.reset_index(drop=True, inplace=True)

    # Standardize the features to have mean = 0 and standard deviation = 1
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)

    # Perform PCA on the standardized features
    pca = PCA()
    pca_result = pca.fit_transform(scaled_features)

    return pca, pca_result, scaler, label_Pathogen, label_Concentration
```

```python
# pathogen_paths = ['./data/SERS Data 1/Concentration study(Salmonella) - modified.xlsx', './data/SERS Data 1/Concnetrations Study(E. coli).xlsx']
# pathogen_paths = ['./data/SERS Data 2/Salmonella data batch 2.xlsx', './data/SERS Data 2/E. coli data batch 2.xlsx']
pathogen_paths = [
    "./data/SERS Data 2/Salmonella (all till Jul 24).xlsx",
    "./data/SERS Data 2/E. coli (all till Jul 24).xlsx",
    "./data/SERS Data 1/Rinsate.xlsx",
]
pathogen_names = ["Salmonella", "E. Coli", "Rinsate"]
combined_samples = load_and_clean_data(pathogen_paths, pathogen_names)

print(combined_samples.columns)
print(combined_samples.index)
```

```python
pca, pca_result, scaler, label_Pathogen, label_Concentration = standardize_and_pca(combined_samples)
print(pca)
print(pca_result.shape)
print(label_Concentration)
```

```python
# Explained variance
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# Scree plot
plt.figure(figsize=(8, 6))
plt.bar(
    range(1, len(explained_variance) + 1),
    explained_variance,
    alpha=0.5,
    align="center",
    label="Individual explained variance",
)
plt.step(
    range(1, len(cumulative_variance) + 1),
    cumulative_variance,
    where="mid",
    label="Cumulative explained variance",
)
plt.ylabel("Explained variance ratio")
plt.xlabel("Principal components")
plt.title("Explained Variance by Principal Components")
plt.legend(loc="best")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.axhline(y=0, color="k", linewidth=0.5)
plt.axvline(x=0, color="k", linewidth=0.5)
plt.xlim(left=0)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


def load_and_clean_data(
    pathogen_paths, pathogen_names, selected_columns, index_column="Raman Shift"
):
    all_pathogen_samples_list = []
    for i, pathogen_path_i in enumerate(pathogen_paths):
        # Load the datasets
        pathogen_i_data = pd.read_excel(pathogen_path_i)

        # Clean column names
        pathogen_i_data.columns = pathogen_i_data.columns.str.strip()

        # Select only the desired columns and the index column
        pathogen_i_data = pathogen_i_data[[index_column] + selected_columns[i]]

        # Set 'Raman Shift' as the index before transposing to keep it as feature names
        pathogen_i_samples = pathogen_i_data.set_index(index_column).T

        # Assign label of pathogen to the samples
        pathogen_i_samples["Pathogen"] = pathogen_names[i]

        # Assign Concentration
        pathogen_i_samples["Concentration"] = pathogen_i_samples.index

        # Add pathogen_i to list
        all_pathogen_samples_list.append(pathogen_i_samples)

    combined_samples = pd.concat(all_pathogen_samples_list)

    return combined_samples


def standardize_and_pca(data):
    # Drop the 'Pathogen' and 'Concentration' columns to isolate features
    features = data.drop(columns=["Pathogen", "Concentration"]).apply(
        pd.to_numeric, errors="coerce"
    )

    # Extract 'Pathogen' and 'Concentration' columns as separate labels
    label_Pathogen = data["Pathogen"]
    label_Concentration = data["Concentration"]

    # Drop any rows with NaN values in the features DataFrame
    features.dropna(inplace=True)

    # Reset indices to ensure clean, sequential indexing
    features.reset_index(drop=True, inplace=True)
    label_Pathogen.reset_index(drop=True, inplace=True)
    label_Concentration.reset_index(drop=True, inplace=True)

    # Standardize the features to have mean = 0 and standard deviation = 1
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)

    # Perform PCA on the standardized features
    pca = PCA()
    pca_result = pca.fit_transform(scaled_features)

    return pca, pca_result, scaler, label_Pathogen, label_Concentration


# Paths to the datasets and the columns to be selected
pathogen_paths = [
    "./data/SERS Data 2/Salmonella data batch 2.xlsx",
    "./data/SERS Data 2/E. Coli data batch 2.xlsx",
]
pathogen_names = ["Salmonella", "E. Coli"]
selected_columns = [["5000 CFU"], ["1213 CFU"]]

# Load and clean data
combined_samples = load_and_clean_data(pathogen_paths, pathogen_names, selected_columns)

print(combined_samples.columns)
print(combined_samples.index)

# Perform PCA
pca, pca_result, scaler, label_Pathogen, label_Concentration = standardize_and_pca(combined_samples)
print(pca)
print(pca_result.shape)
print(label_Concentration)

# Plotting PCA results
plt.figure(figsize=(10, 7))
for pathogen in np.unique(label_Pathogen):
    subset = pca_result[label_Pathogen == pathogen]
    plt.scatter(subset[:, 0], subset[:, 1], label=pathogen, alpha=0.6)
    for i in range(subset.shape[0]):
        plt.text(
            subset[i, 0],
            subset[i, 1],
            label_Concentration[label_Pathogen == pathogen].iloc[i],
            fontsize=8,
            alpha=0.7,
        )

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA: PC1 vs PC2 with Concentration Labels")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
def generate_synthetic_samples_by_mean(df, label):
    synthetic_samples = []
    indices = df.index
    pairs = list(combinations(indices, 2))
    for i, j in pairs:
        mean_sample = (df.loc[i] + df.loc[j]) / 2
        synthetic_samples.append(mean_sample)
    synthetic_df = pd.DataFrame(synthetic_samples, columns=df.columns)
    synthetic_df["Label"] = label
    synthetic_df["Identifier"] = [f"{idx1}_{idx2}_mean" for idx1, idx2 in pairs]
    return synthetic_df
```

```python
def prepare_samples(data, label, synthetic_label=None):
    data["Label"] = label
    data["Identifier"] = data.index
    if synthetic_label:
        synthetic_data = generate_synthetic_samples_by_mean(data, synthetic_label)
        data = pd.concat([data, synthetic_data])
    return data
```

```python
def biplot(score, coeff, labels=None):
    xs = score[:, 0]
    ys = score[:, 1]
    n = coeff.shape[0]
    plt.figure(figsize=(14, 10))
    plt.scatter(xs, ys, c="blue", edgecolors="k")

    for i in range(n):
        plt.arrow(0, 0, coeff[i, 0] * max(xs) * 6, coeff[i, 1] * max(ys) * 6, color="r", alpha=0.5)
        if labels is None:
            plt.text(
                coeff[i, 0] * max(xs) * 5,
                coeff[i, 1] * max(ys) * 5,
                "Var" + str(i + 1),
                color="g",
                ha="center",
                va="center",
            )
        else:
            plt.text(
                coeff[i, 0] * max(xs) * 6,
                coeff[i, 1] * max(ys) * 6,
                labels[i],
                color="g",
                ha="center",
                va="center",
            )

    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()


# Call the biplot function
biplot(pca_result, np.transpose(pca.components_), labels=combined_samples.columns[:-2])
plt.show()
```

```python
def plot_explained_variance(explained_variance):
    cumulative_variance = np.cumsum(explained_variance)
    plt.figure(figsize=(8, 6))
    plt.bar(
        range(1, len(explained_variance) + 1),
        explained_variance,
        alpha=0.5,
        align="center",
        label="Individual explained variance",
    )
    plt.step(
        range(1, len(cumulative_variance) + 1),
        cumulative_variance,
        where="mid",
        label="Cumulative explained variance",
    )
    plt.ylabel("Explained variance ratio")
    plt.xlabel("Principal components")
    plt.title("Explained Variance by Principal Components")
    plt.legend(loc="best")
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.axhline(y=0, color="k", linewidth=0.5)
    plt.axvline(x=0, color="k", linewidth=0.5)
    plt.xlim(left=0)
    plt.show()
```
