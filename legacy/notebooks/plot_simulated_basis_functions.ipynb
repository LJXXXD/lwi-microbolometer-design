```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = "../data/Test 3 - 4 White Powers/Basis functions_4-20um.xlsx"
df = pd.read_excel(file_path)

# Extract wavelength and selected basis functions (0, 2, 4, 6)
wavelength = df.iloc[:, 0]
selected_indices = [0, 2, 4, 6]
basis_functions = df.iloc[:, [i + 1 for i in selected_indices]]

# Estimate peak positions for labels
peak_labels = []
for i in selected_indices:
    curve = df.iloc[:, i + 1]
    peak_wavelength = wavelength[curve.idxmax()]
    peak_labels.append(f"Channel {i + 1} ({peak_wavelength:.0f} µm)")

# Plot
plt.figure(figsize=(6, 3))
for i, col in enumerate(basis_functions.columns):
    plt.plot(wavelength, basis_functions[col], label=peak_labels[i], linewidth=2)

plt.title("Microbolometer Spectral Responses(Simulated)", fontsize=10)
plt.xlabel("Wavelength (µm)", fontsize=9)
plt.ylabel("Absorbance", fontsize=9)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend(fontsize=8)
plt.tight_layout()

# Ensure x and y axes are visible
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(True)
ax.spines["bottom"].set_visible(True)

plt.grid(False)
plt.show()
```
